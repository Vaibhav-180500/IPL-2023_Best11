# IPL 2023 Data Cleaning

```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

## Match Summary

```python
#Importing IPL_Match_Summary csv file

df_match = pd.read_csv('C:/Users/Vaibhav/OneDrive/Desktop/IPL vs BBL/IPL_2023/IPL_Data_Set/IPL_Match_Summary.csv')
df_match.head(1000)
```

```python
#Dropping the unnamed column 

df_match = df_match.drop(df_match.columns[-1], axis=1)
df_match.head()
```

```python
#Renaming column name

df_match.rename({'Match No.':'Match_id'}, axis=1, inplace = True)
df_match.head()
```

```python
#Replacing the values

df_match['Match_id'] = df_match['Match_id'].replace(['Qualifier1', 'Qualifier2'], 'Qualifier')
df_match['Margin'] = df_match['Margin'].fillna('No result')
df_match.tail(30)
```

## Batting Summary

```python
#Importing IPL_Batting_Scorecard csv file

df_batting = pd.read_csv('C:/Users/Vaibhav/OneDrive/Desktop/IPL vs BBL/IPL_2023/IPL_Data_Set/IPL_Batting_Scorecard.csv')
df_batting.head(5)
```

```python
#Renaming column names

df_batting.rename({'selection1_batsmen_name':'batsmen', 'selection1_batsmen_Out_NotOut':'out_notout','selection1_batsmen_Runs':'runs', 'selection1_batsmen_Balls':'balls', 'selection1_batsmen_Fours':'fours', 'selection1_batsmen_Sixes':'sixes', 'selection1_batsmen_Strike_Rate':'sr', 'selection1_batsmen_Team':'team', 'selection1_batsmen_Match_No':'Match_id' }, axis=1, inplace = True)
df_batting.head(20)
```

```python
#Dropping rows that meets the specific condition

to_drop = df_batting[df_batting['runs'].str.contains('Did not bat:', na=False)].index
df_batting = df_batting.drop(to_drop)
df_batting = df_batting.reset_index(drop=True)
df_batting.head(20)
```

```python
#Changing the out_notout column values to out and not out

df_batting['out_notout'] = np.where(df_batting['out_notout'] == 'not out', df_batting['out_notout'], 'out')
df_batting.head(10)
```

```python
#Removing special characters in batsmen column

df_batting['batsmen'] = df_batting['batsmen'].apply(lambda x: x.replace('†', ''))
df_batting['batsmen'] = df_batting['batsmen'].apply(lambda x: x.replace('(c)', ''))

df_batting.head(30)
```

```python
#Matching the Match_id column with the df_match table

df_batting['Match_id'] = df_batting['Match_id'].str.replace(r'^(Q\w+)\s(.*)', r'\1 \2', regex=True)
df_batting['Match_id'] = df_batting['Match_id'].str.replace(r'^(\w+)\s(.*)', r'\1', regex=True)

df_batting.tail(30)
```

```python
#Exporting csv file

#df_batting.to_csv('C:/Users/Vaibhav/OneDrive/Desktop/IPL vs BBL/IPL_2023/IPL_Data_Set_After_Cleaning/Cleaned_IPL_Batting_Summary.csv', index = False)
```

# Bowling Summary
